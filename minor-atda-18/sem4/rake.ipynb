```python
import re
regex = re.compile("[A-Za-zА-Яа-я-.,:–ё]+") 
def words_only(text, regex=regex):
    return " ".join(regex.findall(text))


from pymystem3 import Mystem
m = Mystem()
def lemmatize(text, mystem=m):
    try:
        return "".join(m.lemmatize(text)).strip()  
    except:
        return " "

tokenized_text = words_only(text)
lemmatized_text = lemmatize(tokenized_text)
```

```python
from nltk.corpus import stopwords
mystopwords = stopwords.words('russian') + ['это', 'вне']
with open('./stopwords.txt','w+') as stopwords_file:
    for sw in mystopwords:
        stopwords_file.write(sw + '\n')
```
