## Извлечение ключевых слов по TextRank

```python
text = 'Telegram пока не сдан. Дуров сделал новое заявление в затянувшейся перебранке с Жаровым о статусе свободного интернета в России. Создатель Telegram сказал, что не будет возражать против использования данных компании в реестре «организаторов распространения информации» (ОРИ). Но при этом по-прежнему категорически возражает против слежки за пользователями или выполнения требований пакета Яровой.  Рынок отреагировал на это сообщение мгновенно. Появились новости о том, что Дуров приготовился к капитуляции. Эти настроения кажутся поспешными. Накануне Дуров опроверг информацию о том, что в компании готовят технологию обхода блокировок в «два клика» — по его словам, не нужно принимать написанное в неофициальных каналах мессенджера за корпоративную позицию. Ясно, что в реальности блокировка Telegram в России повлечет за собой потерю массового пользователя, который предпочтет выбирать более простые решения, доступные на рынке, — вроде находящегося пока вне зоны особого внимания властей WhatsApp.   Последнее же заявление, кажется, было сформулировано создателем Telegram намеренно двусмысленно. Дуров не говорит, что он намерен вносить данные компании в реестр Роскомнадзора. Он утверждает лишь, что эти данные и так общеизвестны, и что он не возражает против их внесения в реестр. Кто должен вносить данные, условно, Дуров или Жаров, остается при этом открытым.  Позиция Telegram, таким образом, заключается в том, что в «предоставлении пяти идентификаторов для реестра», на которых настаивает Роскомнадзор, нет никакой проблемы. И говорить о том, что Telegram нарушает в этом отношении российское законодательство тоже нет причины. Вполне можно допускать, что представители Telegram также отправили информацию, дублирующую заявление Дурова, в Роскомнадзор по официальным каналам. Регистрационные данные компании, требующиеся для реестра ОРИ никак не помогут российским властям в деле перлюстрации мессенджера.  Теперь мяч на стороне Роскомнадзора. В ведомстве должны рассказать, удовлетворены ли они, или же мессенджер ждет следующая порция претензий.  Мы бы поставили на второй вариант — возможно, его реализация начнется после небольшой паузы.'
print(text)
```

```python
import re
from nltk.corpus import stopwords
import pymorphy2
prog = re.compile("[А-Яа-яёA-Za-z]+")
l1 = prog.findall(text.lower())
morph = pymorphy2.MorphAnalyzer()
l3 = [morph.parse(token)[0].normal_form for token in l1 if not token in stopwords.words('russian')+['ори']]
text1 = ' '.join(l3)
```

```python
%%time
from gensim.summarization import keywords
kw = keywords(text1)
```

```python
import pandas as pd
rankings = pd.DataFrame({'Text Rank': kw.split('\n')})
rankings.head(20)
```
