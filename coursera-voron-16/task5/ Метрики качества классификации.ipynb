```python
import numpy as np
import pandas as pd

import sklearn.metrics as mtr
```

```python
df = pd.read_csv('classification.csv')
```

```python
df
```

```python
TP = len(df[(df['true'] == 1) & (df['pred'] == 1)])
FP = len(df[(df['true'] == 0) & (df['pred'] == 1)])
FN = len(df[(df['true'] == 1) & (df['pred'] == 0)])
TN = len(df[(df['true'] == 0) & (df['pred'] == 0)])
print(TP, FP, FN, TN)
```

```python
accuracy = mtr.accuracy_score(df['true'], df['pred'])
precision = mtr.precision_score(df['true'], df['pred'])
recall = mtr.recall_score(df['true'], df['pred'])
F1 = mtr.f1_score(df['true'], df['pred'])
print('{:.2f} {:.2f} {:.2f} {:.2f}'.format(accuracy, precision, recall, F1))
```
